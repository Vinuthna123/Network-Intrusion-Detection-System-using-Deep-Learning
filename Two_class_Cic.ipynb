```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
# Loading the Cic Ids dataset dated 2nd March, 2018.
data = pd.read_csv('03-02-2018.csv')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x=data['Label'].value_counts().index, y=data['Label'].value_counts())
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.grid(True)
```

```python
data['Label'].value_counts()
```

```python
## Data Preprocessing
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data.iloc[:,-1] = encoder.fit_transform(data.iloc[:,-1])
```

```python
data['Label'].value_counts()
```

```python
data.describe()
```

```python
## Dropping columns with redundant and null values
data.drop(['Flow Byts/s'],axis=1,inplace=True)
data.drop(['Flow Pkts/s'],axis=1,inplace=True)
```

```python
data.describe()
```

```python
label = data['Label']
```

```python
label
```

```python
data.drop(['Label'],axis=1, inplace=True)
```

```python
data.shape
```

```python
data.drop(['Timestamp'], axis=1, inplace=True)
```

```python
data.head()
data.shape
```

```python
## Separating the features and label
X = data.iloc[:,:76].values
Y = label.values
X[0].shape
Y
```

```python
##Splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
## Normalizing the dataset
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
x_train_normal = normalizer.fit_transform(x_train)
x_test_normal  = normalizer.transform(x_test)
```

```python
x_train_normal
```

```python
x_train_normal.shape
```

```python
import tensorflow as tf
y_train_cat=tf.keras.utils.to_categorical(y_train)
y_test_cat=tf.keras.utils.to_categorical(y_test)
y_train_cat.shape
```

```python
## Building the neural network
from keras.models import Sequential
from keras.layers import Dense, Dropout, Input

model = Sequential() # initializing model
model.add(Input(shape=(76,)))
model.add(Dense(64,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(units=2,activation='softmax'))


model.compile(optimizer='adam',loss='categorical_crossentropy' ,metrics=['accuracy'])
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
## Training the dataset
history = model.fit(x_train_normal,y_train_cat  ,epochs=2)
```

```python
## Testing the model
res = model.evaluate(x_test_normal, y_test_cat, verbose=1)
res
```

```python
## Comparison of true labels and predicted labels using confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score,plot_confusion_matrix

prediction = model.predict_classes(x_test_normal, batch_size=128, verbose=0)
true_labels = np.argmax(y_test_cat, axis=1)
matrix = confusion_matrix(true_labels, prediction)
matrix
```

```python
plt.clf()
plt.figure(figsize=(6, 6))
plt.imshow(matrix, cmap=plt.cm.Wistia)
classNames = ['Normal','Attack']
plt.title('Confusion Matrix - Test Data')
plt.ylabel('True label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
s = [['TN','FP'], ['FN', 'TP']]

for i in range(2):
    for j in range(2):
        plt.text(j,i, str(s[i][j])+" = "+str(matrix[i][j]))
plt.show()
```

```python
## Calculating AUC ROC score
from sklearn.metrics import roc_auc_score,roc_curve
acc = roc_auc_score(true_labels, prediction)
```

```python
acc
```

```python
fpr, tpr, _ = roc_curve(true_labels, prediction)
```

```python
## Plotting the Receiver operating characteristics curve
plt.plot(fpr, tpr, marker='.', label='DeepLearning')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the legend
plt.legend()
# show the plot
plt.show()
```
